```python
import numpy as np

x_array=np.array([[45,85],[50,43],[40,80],[55,42],[200,43],[48,40],[195,41],[43,87],[190,40]])

w1=np.random.randint(40,200, size=2)
w2=np.random.randint(40,200, size=2)
w3=np.random.randint(40,200, size=2)
w_array=[]
d_array=[]
w_array.extend([w1,w2,w3])
print(x_array)  
print(w_array)

def castigator(x_array,w_array,c,ep):
    for e in range(ep):
        for i in range(0,len(x_array)):
            for j in range(0,len(w_array)):
                d=abs(x_array[i][0]-w_array[j][0])+abs(x_array[i][1]-w_array[j][1])
                d_array.append(d)

            d_min=np.min(d_array)
            index=d_array.index(d_min)
            w_array[index]=w_array[index]+c*(x_array[i]-w_array[index])
            d_array.clear()
    print("\nPrototipuri finale:",w_array)    
    
def castigator_afisare_asocieri(x_array,w_array,c,ep):
    for e in range(ep):
        print("\nEpoca",e+1)
        for i in range(0,len(x_array)):
            for j in range(0,len(w_array)):
                d=abs(x_array[i][0]-w_array[j][0])+abs(x_array[i][1]-w_array[j][1])
                d_array.append(d)

            d_min=np.min(d_array)
            index=d_array.index(d_min)
            w_array[index]=w_array[index]+c*(x_array[i]-w_array[index])
            d_array.clear()
            print("\nPattern-ul x"+str(i+1),x_array[i],"este asociat prototipului",w_array[index])
    print("\nPrototipuri finale:",w_array)    
    

print("Functia testata pentru c=1, 5 epoci")
castigator_afisare_asocieri(x_array,w_array,1,5)
print()
print("Functia testata pentru c=0.8, 1500 epoci")  
castigator(x_array,w_array,0.5,1500)
print()
print("Functia testata pentru c=1, 1500 epoci")  
castigator(x_array,w_array,0.5,1500)
```
