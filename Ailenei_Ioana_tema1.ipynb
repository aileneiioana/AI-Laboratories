```python
import numpy as np


def perceptron(x,w,c,x_dorit,nr_epoci):
    semn=[]
    for i in range(len(x)):
        w.append([])
    while(True):
        for i in range(len(x)):  
            net =  np.sum(np.multiply(x[i] ,w[i]))
            semn.append(np.sign(net))
            print("\nnet=",net)
            print("semn",np.sign(net)) 
            if (x_dorit[i]==np.sign(net)):
                w[i+1]=w[i]
                print("Ponderea nu se corecteaza")
            else:
                w[i+1]=w[i]+np.multiply(c*(x_dorit[i]+x_dorit[i]),x[i])
                print("Ponderea se corecteaza")
            print("w",w[i+1])   
        if(x_dorit==semn):
            break
        else: semn.clear()
        w[0]=w[len(w)-1]
            
            
print("Exercitiul 8, pagina 50")
w1=[[0,1,0]]
x1=[[2,1,-1],[0,-1,-1]]
x_dorit1=[-1,1]
c=0.1
perceptron(x1,w1,c,x_dorit1,6)
print()
print()
print("Exemplu curs")
x=[[1,-2,0,-1],[0,1.5,-0.5,-1],[-1,1,0.5,-1]]
w=[[1,-1,0,0.5]]
x_dorit=[-1,-1,1]
perceptron(x,w,c,x_dorit,6)

```
