```python
import numpy as np

x=[[1,-2,1.5,0],[1,-0.5,-2,-1.5],[0,1,-1,1.5]]
w=[1,-1,0,0.5]

y=[[1,-2],[0,1],[2,3],[1,-1]]
q=[1,-1]

def hebb_binar(x,w):
    for i in x:
        print("x= "+str(i))
        net=np.sum(np.multiply(w,i))
        print("net= "+str(net))
        wurm=w+np.multiply(np.sign(net),i)
        print("w_pas_urmator= "+str(wurm))
        w=wurm
        print()
def f(net):
    return 2/(1+np.exp(-net))-1
def hebb_continuu(x,w):
    for i in x:
        print("x= "+str(i))
        net=np.sum(np.multiply(w,i))
        print("net= "+str(net))
        print("f(net)= "+str(f(net)))
        wurm=w+np.multiply(f(net),i)
        print("w_pas_urmator= "+str(wurm))
        w=wurm
        print() 
print("Exercitiul 1:")
print("hebb_functie_activare_binara:")
hebb_binar(x,w)
print("hebb_functie_activare_continua_bipolara_lambda=1:")
hebb_continuu(x,w)
print()
print("Exercitiul 2:")
print("hebb_functie_activare_binara:")
hebb_binar(y,q)
print("hebb_functie_activare_continua_bipolara_lambda=1:")
hebb_continuu(y,q)
print()
```
